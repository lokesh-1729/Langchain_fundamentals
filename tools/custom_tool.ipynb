```python
from langchain_core.tools import BaseTool, StructuredTool, tool
from langchain_core.prompts import PromptTemplate
from langchain_openai import OpenAI
from langchain_core.output_parsers import StrOutputParser
```

```python
@tool
def GetFlightStatus(flight_no:str)-> str:
    """get's fligth status"""
    return f"your flight number is {flight_no} and is about to arrive at 7:00 pm"
```

```python
print(GetFlightStatus.name)
print(GetFlightStatus.description)
print(GetFlightStatus.args)
```

```python
GetFlightStatus
```

```python
prompt=PromptTemplate.from_template("based on the context:{context},\n answer the query:{query} about flight in one word only if you find the details in the context")
```

```python
llm=OpenAI()
```

```python
flight="EK524"
context=GetFlightStatus.run(flight)
```

```python
output_parser=StrOutputParser()
```

```python
chain=prompt|llm|output_parser
```

```python
chain.invoke({"context":context,"query":"arrival_time","flight":flight})
```
