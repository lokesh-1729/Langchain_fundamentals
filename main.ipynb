```python
from langchain_core.prompts import FewShotChatMessagePromptTemplate,ChatPromptTemplate
```

```python
examples=[
    {"input":"India","output":"aidnI"},
    {"input":"canada","output":"adanac"},
    {"input":"china","output":"anihc"},
]
```

```python
example_prompt=ChatPromptTemplate.from_messages([
    ("human","{input}"),
    ("ai","{output}"),
])
```

```python
few_shot_prompt=FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=examples
)
```

```python
print(few_shot_prompt.format())
```

```python
p_template=ChatPromptTemplate.from_messages([
    ("system","you are a linguistic specialist"),
    few_shot_prompt,
    ("human","{input}"),
])
```

```python
p_template
```

```python
prompt=p_template.format_messages(input="Brazil")
```

```python
prompt
```

```python
res=model.invoke(prompt)
```

```python
res.content
```

```python
#llm always return text
from langchain_core.output_parsers import CommaSeparatedListOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_openai import OpenAI
from langchain_core.output_parsers.list import ListOutputParser 
from dotenv import load_dotenv
load_dotenv()
```

```python
llm=OpenAI()
```

```python
prompt=PromptTemplate(
    template="List 3 {things}",
    input_variables=["things"]
)
```

```python
print(llm.invoke(input=prompt.format(things="countries that play cricket in world cup")))
```

```python
output_parser=CommaSeparatedListOutputParser()
format_instructions=output_parser.get_format_instructions()
```

```python
print(format_instructions)
```

```python
prompt=PromptTemplate(
    template="List 3{things}.\n {format_instructions}",
    input_variables=["things"],
    partial_variables={"format_instructions":format_instructions}
)
```

```python
final_prompt=prompt.format(things="countries that play cricket in world cup")
```

```python
final_prompt
```

```python
output=llm.invoke(input=final_prompt)
```

```python
print(output.strip())
```

```python
things=output_parser.parse(output)
print(things)
```
