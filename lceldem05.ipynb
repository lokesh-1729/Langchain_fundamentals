```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI
from langchain_core.runnables import RunnablePassthrough
```

```python
title=(
    ChatPromptTemplate.from_template("generate a impactful title for {input}")|
    ChatOpenAI()
    |StrOutputParser()
    |{"title":RunnablePassthrough()}
)
```

```python
outline=(
    ChatPromptTemplate.from_template(
        "generate a detaile doutline for {title}"
    )
    |ChatOpenAI()
    |StrOutputParser()
    |{"outline":RunnablePassthrough()}
)
```

```python
blog=(
    ChatPromptTemplate.from_template(
        "generate a 200 word blog post based on the outlines{outline}"
    )
    |ChatOpenAI()
    |StrOutputParser()
)
```

```python
summary=(
    ChatPromptTemplate.from_template(
        "generate a 2summary for the post{blog}"
    )
    |ChatOpenAI()
    |StrOutputParser()
)
```

```python
content_chain=title|outline|blog|summary
```

```python
content_chain.invoke({"input":"the impact of ai on jobs"})
```

```python
from langchain_core.runnables import RunnableLambda
```

```python
def get_len(text):
    print(text)
    return len(text)
```

```python
def to_titlecase(text):
    return text.title()
```

```python
new_prompt=ChatPromptTemplate.from_template("give me a one-line description of {topic}")
model=ChatOpenAI()
output_parser=StrOutputParser()
chain=new_prompt|model|output_parser|RunnableLambda(to_titlecase)|RunnableLambda(get_len)
```

```python
chain.invoke({"topic":"AI"})
```
